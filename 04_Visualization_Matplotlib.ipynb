# Data Visualization: Creating graphs using Matplotlib

In this course, we will learn how to perform data visualization using the Matplotlib library. First, we will understand how do the figures and axes work in Matplotlib. Then, we will learn how to change colors, styles, annotations, and others. 

Note that Matplotlib also serves as a baseline for other popular libraries, such as Seaborn and even Pandas plotting. Thus, what we will learn here also serves for customizing figures using these packages. 

# Creating our first visualizations

First, let's import our dataset. This dataset stores a time series for different climate features:

```python
import pandas as pd

climate = pd.read_csv('monitoramento_tempo.csv')
climate
```

Our dataset has 45253. Let's understand it a little better:

```python
climate.info( )
```

So, we have 45253 entries, and there is no null value. Also, our "date" column is not of type datetime. Let's adjust this:

```python
climate.data = pd.to_datetime(climate.data)
```

Nice! Now, let's start visualizing our dataset. Thus, we will use the matplotlib library:

```python
import matplotlib.pyplot as plt

plt.plot(climate['data'], climate['temperatura'])
```

Nice! We have just created our first plot. However, it is hard to understand and visualize it. Let's increase the size of our graph:

```python
plt.figure(figsize = (15, 8))                       # Creating a figure object and changing the size of our figure

plt.plot(climate['data'], climate['temperatura'])
```

Great! Now, let's add a title to our graph:

```python
plt.figure(figsize = (15, 8))                       # Creating a figure object and changing the size of our figure

plt.plot(climate['data'], climate['temperatura'])

plt.title("Temperature vs Date")                    # Setting a title
```

Ok! Let's explore some other basic features from matplotlib:

```python
fig = plt.figure(figsize = (12, 5))                          # Creating a figure object and changing the size of our figure
fig_axis = fig.add_axes([0, 0, 1, 1])                        # Creating figure axis

fig_axis.plot(climate['data'], climate['temperatura'], color = 'green')

fig_axis.set_title("Temperature at each day", fontsize = 18) # Setting a title
fig_axis.set_xlabel('Year', fontsize = 12)                   # Setting xlabel
fig_axis.set_ylabel('Temperature', fontsize = 12)            # Setting ylabel
fig_axis.legend(['Temperature'], loc = 'lower right')        # Creating a legend
```

Note that our graph looks much better now. Note that we have created two objects: one for the figure (fig), and one for the axis (fig_axis). These objects will help us to later manipulate many customization options in Matplotlib.

Also, we have changed the color of our graph. To see which colors we can use in Matplotlib, we can see the following sheet:

# Customizing our visualization

Now, let's see other options to customize our plot. First, we can change the line width using the **lw** parameter:

```python
fig = plt.figure(figsize = (12, 5))                          # Creating a figure object and changing the size of our figure
fig_axis = fig.add_axes([0, 0, 1, 1])                        # Creating figure axis

fig_axis.plot(climate['data'], climate['temperatura'], color = 'green', lw = 0.5)

fig_axis.set_title("Temperature at each day", fontsize = 18) # Setting a title
fig_axis.set_xlabel('Year', fontsize = 12)                   # Setting xlabel
fig_axis.set_ylabel('Temperature', fontsize = 12)            # Setting ylabel
fig_axis.legend(['Temperature'], loc = 'lower right')        # Creating a legend
```

It seems that a thinner line helps us in this case. Other option is to use the **ls** parameter to change the linestyle:

```python
fig = plt.figure(figsize = (12, 5))                          # Creating a figure object and changing the size of our figure
fig_axis = fig.add_axes([0, 0, 1, 1])                        # Creating figure axis

fig_axis.plot(climate['data'], climate['temperatura'], color = 'green', ls = 'dotted')

fig_axis.set_title("Temperature at each day", fontsize = 18) # Setting a title
fig_axis.set_xlabel('Year', fontsize = 12)                   # Setting xlabel
fig_axis.set_ylabel('Temperature', fontsize = 12)            # Setting ylabel
fig_axis.legend(['Temperature'], loc = 'lower right')        # Creating a legend
```

The option 'dotted' makes the line dotted. However, this linestyle does not help in our case. Another option is to add a marker (as in *marker = 'o'*), but, since we have too much data, this option also will not help us.

See that our feature goes from abour 2013 to 2018. We can change the limits of our plot using:

```python
import datetime

fig = plt.figure(figsize = (12, 5))                          # Creating a figure object and changing the size of our figure
fig_axis = fig.add_axes([0, 0, 1, 1])                        # Creating figure axis

fig_axis.plot(climate['data'], climate['temperatura'], color = 'green', lw = 1)

fig_axis.set_xlim(datetime.datetime(2014, 1, 1), datetime.datetime(2015, 1, 1))

fig_axis.set_title("Temperature at each day", fontsize = 18) # Setting a title
fig_axis.set_xlabel('Month', fontsize = 12)                  # Setting xlabel
fig_axis.set_ylabel('Temperature', fontsize = 12)            # Setting ylabel
fig_axis.legend(['Temperature'], loc = 'lower right')        # Creating a legend
```

Now, we are only seeing the time series for one year: 2014. See that the temperature decreases in the year extremes, and increases in the middle of the year.  

To make it easier to understand the temperature in each portion of our plot, we can add a grid:

```python
fig = plt.figure(figsize = (12, 5))                          # Creating a figure object and changing the size of our figure
fig_axis = fig.add_axes([0, 0, 1, 1])                        # Creating figure axis

fig_axis.plot(climate['data'], climate['temperatura'], color = 'green', lw = 1)

fig_axis.set_xlim(datetime.datetime(2014, 1, 1), datetime.datetime(2015, 1, 1))

fig_axis.set_title("Temperature at each day", fontsize = 18) # Setting a title
fig_axis.set_xlabel('Month', fontsize = 12)                  # Setting xlabel
fig_axis.set_ylabel('Temperature', fontsize = 12)            # Setting ylabel
fig_axis.legend(['Temperature'], loc = 'lower right')        # Creating a legend
fig_axis.grid( )
```

## Adding another visualization inside our figure

Note that, when we changed the limits of our graph, we focused on a small portion of the entire dataset. What if we also want to show, in the same figure, the entire time series, but in a smaller size, so that one knows that it comes from a bigger time series?

Before, we created an object for the figure, and another for the figure axis. Manipulating these classes, we are able to create different subplots. Let's show an example:

```python
fig = plt.figure(figsize = (12, 5))                          # Creating a figure object and changing the size of our figure
fig_axis  = fig.add_axes([0, 0, 1, 1])                       # Creating figure axis
fig_axis2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])               # Creating a second axis
```

Note that here we have two axis in the same figure! Now, let's add a figure to these axis:

```python
fig = plt.figure(figsize = (12, 5))                          # Creating a figure object and changing the size of our figure

fig_axis  = fig.add_axes([0, 0, 1, 1])                       # Creating figure axis
fig_axis2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])               # Creating a second axis

# FIRST AXIS

fig_axis.grid( )
fig_axis = fig.add_axes([0, 0, 1, 1])                        # Creating figure axis

fig_axis.plot(climate['data'], climate['temperatura'], color = 'green', lw = 1)

fig_axis.set_xlim(datetime.datetime(2014, 1, 1), datetime.datetime(2015, 1, 1))

fig_axis.set_title("Temperature at each day", fontsize = 18, pad = 20) # Setting a title
fig_axis.set_xlabel('Month', fontsize = 12)                  # Setting xlabel
fig_axis.set_ylabel('Temperature', fontsize = 12)                      # Setting ylabel
fig_axis.legend(['Temperature'], loc = 'best')                         # Creating a legend
```

Nice! We have added a figure to one of our axis, and now we can add a figure to the other axis. However, note that the axis is in a very bad position. We can change the axis position changing the values when instancing the object. Thus, let's do:

```python
fig = plt.figure(figsize = (16, 8))                          # Creating a figure object and changing the size of our figure

fig_axis  = fig.add_axes([0, 0, 1, 1])                       # Creating figure axis
fig_axis2 = fig.add_axes([0.05, 0.65, 0.3, 0.3])               # Creating a second axis

# FIRST AXIS

fig_axis.grid( )

fig_axis.plot(climate['data'], climate['temperatura'], color = 'green', lw = 1)

fig_axis.set_xlim(datetime.datetime(2014, 1, 1), datetime.datetime(2015, 1, 1))

fig_axis.set_title("Temperature at each day", fontsize = 18, pad = 20) # Setting a title
fig_axis.set_xlabel('Month', fontsize = 12)                  # Setting xlabel
fig_axis.set_ylabel('Temperature', fontsize = 12)                      # Setting ylabel
fig_axis.legend(['Temperature'], loc = 'best')                         # Creating a legend

# SECOND AXIS

fig_axis2.plot(climate['data'], climate['temperatura'], color = 'blue', lw = 0.3)
fig_axis2.set_xlabel('Year', fontsize = 8)                             # Setting xlabel
fig_axis2.set_ylabel(' ')                                              # Setting ylabel
```

Nice! Our graph looks very good. Note that, since our second axis is small, we simplified it, so that it does not look too crowded. Finally, let's just do something to make it easier to understand where our data from the main axis is on the second axis:

```python
fig = plt.figure(figsize = (16, 8))                          # Creating a figure object and changing the size of our figure

fig_axis  = fig.add_axes([0, 0, 1, 1])                       # Creating figure axis
fig_axis2 = fig.add_axes([0.03, 0.68, 0.3, 0.3])               # Creating a second axis

# FIRST AXIS

fig_axis.grid( )

fig_axis.plot(climate['data'], climate['temperatura'], color = 'green', lw = 1)

fig_axis.set_xlim(datetime.datetime(2014, 1, 1), datetime.datetime(2015, 1, 1))
fig_axis.set_ylim(260, 330)

fig_axis.set_title("Temperature in 2014", fontsize = 18, pad = 20)     # Setting a title
fig_axis.set_xlabel('Month', fontsize = 12)                            # Setting xlabel
fig_axis.set_ylabel('Temperature', fontsize = 12)                      # Setting ylabel
fig_axis.legend(['Temperature'], loc = 'best')                         # Creating a legend

# SECOND AXIS

climate_filter = climate.query("data > datetime.datetime(2014, 1, 1) and data < datetime.datetime(2015, 1, 1)")

fig_axis2.set_ylim(260, 330)

fig_axis2.plot(climate['data'], climate['temperatura'], color = 'blue', lw = 0.2)
fig_axis2.plot(climate_filter['data'], climate_filter['temperatura'], color = 'green', lw = 0.2)
fig_axis2.set_xlabel('Year', fontsize = 8)                             # Setting xlabel
fig_axis2.set_ylabel(' ')                                              # Setting ylabel
```

Nice! We have filtered our data and ploted a second time series in axis 2, now showing the location of the main time series in the total series. 

We have also changed other minor things to improve our visualization, such as line width, location of the second axis and axis limits.

# Anotations and more graphs

Our time series for the temperature looks very good. Now, let's try to add some annotations, to better pass a message to the team. So, let's add horizontal lines delimiting a threshold, and let's add some markers for the maximum and minimum points.

```python
fig = plt.figure(figsize = (16, 8))
axis1 = fig.add_axes([0, 0, 1, 1])

axis1.plot(climate['data'], climate['temperatura'], color = 'g')

axis1.set_xlim(datetime.datetime(2014, 1, 1), datetime.datetime(2015, 1, 1))
axis1.set_ylim(260, 330)
axis1.set_title('Temperature in 2014', fontsize = 20, pad = 20)
axis1.set_xlabel('Date', fontsize = 16)
axis1.set_ylabel('Temperature', fontsize = 16)
axis1.legend(['Temperature'], loc = 'best', fontsize = 16)

# Adding our thresholds

climate['year'] = climate.data.apply(lambda x : x.year)
max_temperature_2014 = max(climate[climate.year == 2014].temperatura)
min_temperature_2014 = min(climate[climate.year == 2014].temperatura)
axis1.axhline(max_temperature_2014, color = 'k', linestyle = '--')
axis1.axhline(min_temperature_2014, color = 'k', linestyle = '--')
```

Nice! We have added thresholds for the maximum and minimum values. Now, let's add an annotation for the maximum temperature:

```python
fig = plt.figure(figsize = (16, 8))
axis1 = fig.add_axes([0, 0, 1, 1])

axis1.plot(climate['data'], climate['temperatura'], color = 'g')

axis1.set_xlim(datetime.datetime(2014, 1, 1), datetime.datetime(2015, 1, 1))
axis1.set_ylim(260, 330)
axis1.set_title('Temperature in 2014', fontsize = 20, pad = 20)
axis1.set_xlabel('Date', fontsize = 16)
axis1.set_ylabel('Temperature', fontsize = 16)
axis1.legend(['Temperature'], loc = 'best', fontsize = 16)

# Adding our thresholds

climate['year'] = climate.data.apply(lambda x : x.year)
max_temperature_2014 = max(climate[climate.year == 2014].temperatura)
min_temperature_2014 = min(climate[climate.year == 2014].temperatura)

axis1.axhline(max_temperature_2014, color = 'k', linestyle = '--')
axis1.axhline(min_temperature_2014, color = 'k', linestyle = '--')

# Adding an annotation

xmax = climate['data'][climate.query('year == 2014').temperatura.idxmax( )]
ymax = max_temperature_2014

axis1.annotate('Maximum temperature', xy = (xmax, ymax), fontsize = 14)
```

Note that our annotation was added exactly at the position of the maximum temperature in 2014. However, it looks very bad in the graph. Let's try to improve this by adding an arrow:

```python
fig = plt.figure(figsize = (16, 8))
axis1 = fig.add_axes([0, 0, 1, 1])

axis1.plot(climate['data'], climate['temperatura'], color = 'g')

axis1.set_xlim(datetime.datetime(2014, 1, 1), datetime.datetime(2015, 1, 1))
axis1.set_ylim(260, 330)
axis1.set_title('Temperature in 2014', fontsize = 20, pad = 20)
axis1.set_xlabel('Date', fontsize = 16)
axis1.set_ylabel('Temperature', fontsize = 16)
axis1.legend(['Temperature'], loc = 'best', fontsize = 16)

# Adding our thresholds

climate['year'] = climate.data.apply(lambda x : x.year)
max_temperature_2014 = max(climate[climate.year == 2014].temperatura)
min_temperature_2014 = min(climate[climate.year == 2014].temperatura)

axis1.axhline(max_temperature_2014, color = 'k', linestyle = '--')
axis1.axhline(min_temperature_2014, color = 'k', linestyle = '--')

# Adding an annotation with an arrow

# Arrow end
x1 = climate['data'][climate.query('year == 2014').temperatura.idxmax( )]
y1 = max_temperature_2014

# Arrow origin
x2 = climate['data'][climate.query('year == 2014').temperatura.idxmax( ) - 800]
y2 = max_temperature_2014 + 3

axis1.annotate('Maximum', xy = (x1, y1), fontsize = 14, xytext = (x2, y2), arrowprops = {'facecolor' : 'k', 'width' : 1})
```

# Visualizing different data

Now, let's try to perform our types of graphs using different a different dataset:

```python
iris = pd.read_csv('iris.csv')
```

```python
iris.head( )
```

In our dataset, we have information about the sepals and petals of some flowers from 3 species: iris-setosa, iris-versicolor and iris-virginica.

```python
iris['espécie'].unique( )
```

Let's start to understand our data. Is the sepal length related to its width?

```python
fig = plt.figure(figsize = (12, 6))
axis = fig.add_axes([0, 0, 1, 1])

axis.scatter(iris['comprimento_sépala'], iris['largura_sépala'])
axis.set_title("Scatter plot", fontsize = 18, pad = 15)
axis.set_xlabel('Sepal length', fontsize = 14)
axis.set_ylabel('Sepal width', fontsize = 14)
axis.tick_params(labelsize = 12)
```

It is hard to see a clear pattern in this data. However, data seems to be divided in groups. Maybe these groups depend on the species? Let's test it:

```python
fig = plt.figure(figsize = (12, 6))
axis = fig.add_axes([0, 0, 1, 1])

colors  = {'Iris-setosa' : 'r', 'Iris-versicolor' : 'b', 'Iris-virginica' : 'g'}
markers = {'Iris-setosa' : 'x', 'Iris-versicolor' : 'o', 'Iris-virginica' : 's'}

for species in iris['espécie'].unique( ):
  tmp = iris[iris['espécie'] == species]
  axis.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'], color = colors[species], marker = markers[species])

axis.legend(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'], fontsize = 14)
axis.set_title("Scatter plot", fontsize = 18, pad = 15)
axis.set_xlabel('Sepal length', fontsize = 14)
axis.set_ylabel('Sepal width', fontsize = 14)
axis.tick_params(labelsize = 12)
```

Nice! Here, we have a clear pattern in our data. Iris-setosa presents a greater width-per-length ratio, and it has a very distinct pattern. Iris-virginica, on the other hand, presents a greater length-to-width ratio. However, in some regions, it is hard to identify if the species is iris-virginica or iris-versicolor only looking at the sepal width and length.

So, now, let's try to understand the distribution of our data. For that end, let's use a boxplot:

```python
fig = plt.figure(figsize = (5, 4))
axis = fig.add_axes([0, 0, 1, 1])

axis.boxplot(iris['largura_pétala'])
axis.set_title('Boxplot', fontsize = 15, pad = 10)
axis.set_xticklabels(['Petal width'])
```

Ok, we have drawn a boxplot for the petal width. Now, let's add the other features (except species):

```python
fig = plt.figure(figsize = (8, 4))
axis = fig.add_axes([0, 0, 1, 1])

tmp = iris.drop('espécie', axis = 1)
axis.boxplot(tmp.values)
axis.set_xticklabels(tmp.columns)
axis.set_title('Boxplot', fontsize = 15, pad = 10)
```

Nice! We managed to draw a boxplot for each feature, and define the label names as the name of the columns from our dataframe. Let's try to make our plot more pleasing for the viewer:

```python
fig = plt.figure(figsize = (8, 4))
axis = fig.add_axes([0, 0, 1, 1])

colors = ['mediumpurple', 'lightblue', 'bisque', 'palegreen']

tmp = iris.drop('espécie', axis = 1)
box_plot = axis.boxplot(tmp.values, patch_artist = True)
axis.set_xticklabels(tmp.columns)
axis.set_title('Boxplot', fontsize = 15, pad = 10)

for box, color in zip(box_plot['boxes'], colors):
  box.set(color = color)
```

Nice! Now, let's look a little more to our boxplots. Our boxplots show how the distribution of data is given, showing the median, and the first and third quartiles in a box. Also, it shows some outliers (according to a given outlier detection rule). These outliers are shown by the 'o' sign. 

We can change how these outliers are shown in our plot:

```python
fig = plt.figure(figsize = (8, 4))
axis = fig.add_axes([0, 0, 1, 1])

colors = ['mediumpurple', 'lightblue', 'bisque', 'palegreen']

tmp = iris.drop('espécie', axis = 1)
box_plot = axis.boxplot(tmp.values, patch_artist = True)
axis.set_xticklabels(tmp.columns)
axis.set_title('Boxplot', fontsize = 15, pad = 10)

for box, color in zip(box_plot['boxes'], colors):
  box.set(color = color)

for outlier in box_plot['fliers']:
  outlier.set(marker = 'x', markersize = 8)
```

Another way to visualize data distribution is by using an histogram. To plot a histogram, we can simply do:

```python
fig = plt.figure(figsize = (8, 4))
axis = fig.add_axes([0, 0, 1, 1])

axis.hist(iris['comprimento_pétala'])
axis.set_title('Histogram', fontsize = 15, pad = 10)
axis.set_xlabel('Petal length')
axis.grid( )
```

Nice. Using histograms, we can also change the number of bins, and change it from a count histogram to a density plot:

```python
fig = plt.figure(figsize = (8, 4))
axis = fig.add_axes([0, 0, 1, 1])

axis.hist(iris['comprimento_pétala'], bins = 20, density = True)
axis.set_title('Histogram', fontsize = 15, pad = 10)
axis.set_xlabel('Petal length')
axis.grid( )
```

The density plot it very interesting when we are working with probability distributions. 

Ok, we are seeing our distribution through some bars represented by the histogram. Is there any way to visualize, in the same plot, some central tendency measures, such as the mean and the median?

For that end, we can use annotations and vertical lines:

```python
fig = plt.figure(figsize = (8, 4))
axis = fig.add_axes([0, 0, 1, 1])

mu, sigma, med = iris['comprimento_pétala'].mean( ), iris['comprimento_pétala'].std( ), iris['comprimento_pétala'].median( )

axis.hist(iris['comprimento_pétala'], bins = 20)
axis.set_title('Histogram', fontsize = 15, pad = 10)
axis.set_xlabel('Petal length')
axis.grid( )

axis.annotate('$\mu = {0:.2f}$ \n $\sigma = {1:.2f}$'.format(mu, sigma), xy = (6, 20), fontsize = 20)
axis.axvline(mu, color = 'k', linestyle = '--')
axis.annotate('$\mu$', xy = (mu-0.3, 28), fontsize = 20)

axis.axvline(med, color = 'g', linestyle = '--')
axis.annotate('median', xy = (med+0.05, 31), fontsize = 20)
```

Nice! This way, our plot is able to pass much more information. Now, let's take a closer look to our histogram. Note that, as in the scatter plot from before, we can see that there are groups in our data. Let's see if we can separate these groups in different histograms. First, let's try to visualize a histogram only for iris-versicolor:

Now, let's draw these figures in the same plot. We can make this by importing our saved figures:
